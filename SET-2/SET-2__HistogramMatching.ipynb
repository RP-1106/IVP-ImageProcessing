```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
tif_image = cv2.imread('blur.tif', cv2.IMREAD_UNCHANGED)

# Convert the image to .jpg format
jpg_image = cv2.cvtColor(tif_image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
cv2.imwrite('image.jpg', jpg_image, [int(cv2.IMWRITE_JPEG_QUALITY), 100])
```

```python
image = cv2.imread('image.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
blurred = cv2.GaussianBlur(image, (5, 5), 0)
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(blurred, cmap='gray')
plt.title('Blurred Image')

plt.subplot(1, 2, 2)
plt.hist(blurred.ravel(), 256, [0, 256])
plt.title('Histogram of Blurred Image')
```

```python
equalized = cv2.equalizeHist(blurred)
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(equalized, cmap='gray')
plt.title('Equalized Image')

plt.subplot(1, 2, 2)
plt.hist(equalized.ravel(), 256, [0, 256])
plt.title('Histogram of Equalized Image')

plt.show()
```
