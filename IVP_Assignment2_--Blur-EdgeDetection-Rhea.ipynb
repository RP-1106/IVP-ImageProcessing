```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread('image_new.jpeg')

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

#defines the lower limit of white color
lower_white = np.array([200, 200, 200])
#defines the upper limit of white color
upper_white = np.array([255, 255, 255])

'''pixels within the specified color range become white (255), 
   and pixels outside the range become black (0)'''
mask = cv2.inRange(image, lower_white, upper_white)

'''changes the color of the pixels in the original image where the
mask is non-zero (white) to black. It sets the BGR values of those
pixels to [0, 0, 0], resulting in black color.'''
image[mask > 0] = [0, 0, 0]
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Modified Image')
plt.axis('off')
plt.show()
```
