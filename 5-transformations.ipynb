```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('img.jpeg')
```

# Original Image

```python
plt.imshow(img)
plt.show()
```

# TRANSFORMATIONS

```python
def vertical_reflection(image):
    height, width, channels = image.shape
    reflected_image = np.zeros((height, width, channels), dtype=np.uint8)
    for j in range(width):
        reflected_image[:, j] = image[:, width - j - 1]
    return reflected_image

reflected_img = vertical_reflection(img)
plt.imshow(reflected_img)
plt.show()
```

```python
def horizontal_reflection(image):
    height, width, channels = image.shape
    reflected_image = np.zeros((height, width, channels), dtype=np.uint8)
    for i in range(height):
        reflected_image[i, :] = image[height - i - 1, :]
    return reflected_image

reflected_img = horizontal_reflection(img)
plt.imshow(reflected_img)
plt.show()
```

```python
def scale_image(image, scale_x, scale_y):
    scaling_matrix = np.float32([[scale_x, 0, 0], [0, scale_y, 0]])
    img_scaled = cv2.warpAffine(image, scaling_matrix, (image.shape[1], image.shape[0]))
    return img_scaled

scale_x = 1.5  
scale_y = 0.8 

scaled_img = scale_image(img, scale_x, scale_y)
plt.imshow(scaled_img)
plt.show()
```

```python
def rotate_image(image, angle):
    height, width = image.shape[:2]
    rotation_matrix = cv2.getRotationMatrix2D((width / 2, height / 2), angle, 1)
    rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height))
    return rotated_image

rotation_angle = 45 #in degrees
rotated_image = rotate_image(img, rotation_angle)
plt.imshow(rotated_image)
plt.show()
```

```python
def translate_image(image, tx, ty):
    translation_matrix = np.float32([[1, 0, tx], [0, 1, ty]])
    img_translation = cv2.warpAffine(image, translation_matrix, (image.shape[1], image.shape[0]))
    return img_translation

tx = 150
ty = -30

translated_img = translate_image(img, tx, ty)
plt.imshow(translated_img)
plt.show()
```
